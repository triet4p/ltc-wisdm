```python
import numpy as np
from ltc_wisdm.config import PROCESSED_DATA_DIR

print("Đang đọc các file đã lưu...")
X_train = np.load(os.path.join(PROCESSED_DATA_DIR, 'X_train.npy'))
y_train = np.load(os.path.join(PROCESSED_DATA_DIR, 'y_train.npy'))
X_test = np.load(os.path.join(PROCESSED_DATA_DIR, 'X_test.npy'))
y_test = np.load(os.path.join(PROCESSED_DATA_DIR, 'y_test.npy'))

# --- Kiểm tra Shape và Dtype ---
print("\n--- Kiểm tra Shape và Dtype ---")
print(f"X_train shape: {X_train.shape}, dtype: {X_train.dtype}")
print(f"y_train shape: {y_train.shape}, dtype: {y_train.dtype}")
print(f"X_test shape: {X_test.shape}, dtype: {X_test.dtype}")
print(f"y_test shape: {y_test.shape}, dtype: {y_test.dtype}")

# --- Kiểm tra nội dung của một vài mẫu ---
print("\n--- Kiểm tra nội dung mẫu đầu tiên ---")
# Lấy ra cửa sổ đầu tiên và nhãn tương ứng
first_window = X_train[0]
first_label = y_train[0]

print("Cửa sổ đầu tiên (5 dòng đầu):")
print(first_window[:5, :]) # In ra 5 bước thời gian đầu tiên
print(f"\nNhãn của cửa sổ đầu tiên: {first_label}")

# --- Kiểm tra xem dữ liệu đã được chuẩn hóa chưa ---
print("\n--- Kiểm tra kết quả chuẩn hóa ---")
# Tính mean và std trên toàn bộ tập train
# Dữ liệu đã được chuẩn hóa nên mean sẽ rất gần 0 và std sẽ rất gần 1
mean_val = np.mean(X_train)
std_val = np.std(X_train)
print(f"Mean của X_train: {mean_val:.4f}")
print(f"Standard Deviation của X_train: {std_val:.4f}")
```
